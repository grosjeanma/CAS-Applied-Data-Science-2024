```python
import sqlite3

conn = sqlite3.connect("university.db")
```

Exercise 1.2:
- list all tables
- list all properties of all tables

```python
cursor = conn.cursor()

cursor.execute("select name from sqlite_master where type='table'")
tables = cursor.fetchall()

print("Tables in the database:")
for table in tables:
    print(f"\n{table[0]}")
    cursor.execute(f"PRAGMA table_info({table[0]})")
    columns = cursor.fetchall()
    for column in columns:
        print(column)
```

Exercise 1.3:
- List everything from every tabe

```python
cursor = conn.cursor()

cursor.execute("SELECT name FROM sqlite_master WHERE type='table'")
tables = cursor.fetchall()

print("Tables and data in the database:")
for table in tables:
    print(f"\n{table[0]}")
    cursor.execute(f"select * from {table[0]}")
    columns = cursor.fetchall()
    for column in columns:
        print(column)
```

Exercise 2.2: How many instructor names start with ’E’?


```python
cursor = conn.cursor()

cursor.execute("Select * from instructor where name like 'E%'")
columns = cursor.fetchall()
for column in columns:
    print(column)
```

Exercise 2.3: Find the names of all instructors that have taught in the Watson-building.



```python
cursor = conn.cursor()

cursor.execute("Select * from instructor i join department d on i.dept_name = d.dept_name where d.building = 'Watson'")
columns = cursor.fetchall()
for column in columns:
    print(column)
```
